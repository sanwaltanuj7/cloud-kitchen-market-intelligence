```python
import pandas as pd
```

```python
df=pd.read_csv('raw_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
before_count = len(df)

print("Records Before Cleaning:", before_count)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df['Number of Reviews'] = df['Number of Reviews'].fillna('Not Available')
```

```python
df['Delivery Time'] = df['Delivery Time'].fillna('Not Available')
```

```python
df['Rating'] = df['Rating'].fillna('Not Rated')
```

```python
df['Cuisine(s)'] = df['Cuisine(s)'].str.replace('BBQ','Barbecue')
df['Cuisine(s)'] = df['Cuisine(s)'].str.replace('Coffee','Cafe')
df['Cuisine(s)'] = df['Cuisine(s)'].str.replace('Indo-Chinese','Chinese')
df['Cuisine(s)'] = df['Cuisine(s)'].str.replace('Cake','Bakery')
```

```python
df['Cuisine(s)'] = df['Cuisine(s)'].str.strip()

df['Restaurant Name'] = df['Restaurant Name'].str.strip()
```

```python
df['Type'] = df['Type'].replace({
    'Dine In':'Dine-In Restaurant',
    'Restaurant':'Dine-In Restaurant'
})
```

```python
df.isnull().sum()
```

```python
after_count = len(df)

print("Records After Cleaning:", after_count)
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
